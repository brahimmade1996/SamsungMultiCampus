> get 방식
- url로 모든 것이 넘어가는 방식(http:// ~~~~ search keyword
- url이 너무 길면 잘라버림
- url; 계층을 만들어서 설계(naver.com/최상위계층(store)/ 하위 계층(바지, 상의 등)/상품이름(123))

> post 방식
- url에 keyword, 파라미터를 담아서 보내지 않고, body에 담아서 보냄.
- 데이터의 양에 상관없이 보낼 수 있음.


#### 통신방식 
- TCP 통신규약(프로토콜)로 통신. binary로 데이터를 주고 받음.
- TCP + HTTP: text로 데이터를 주고 받음. 틀이 없음.
- REST API: URL(리소스), body, method(행위)를 이용하여 통신

> method: CRUD(Create, Retrive, Update, Delete)
