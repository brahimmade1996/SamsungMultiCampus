# 1. verture machine vs containers

## 가상머신
> 왜 나왔을가?
- 하드웨어 사양이 점점 좋아지는데,
- 아파치 웹 서버를 여러개 띄웠는데, 이벤트로 하나가 폭주하면
- 하나가 나머지 서버에 영향을 주므로, 격리하기 위해서 사용
- 나머지는 살아남고 폭주한 서버만 죽도록

> 문제
- hypervisor(os, 리눅스) 위에 guest os가 더 들어감 -> overhead가 생김(5~10%문제)
- os가 2개이므로 생기는 overhead문제를 해결하기 위해 컨테이너 도입

## 컨테이너
- os위에 간단한 컨테이너 엔진을 얹어서 사용
- guest os가 application 레벨로 바뀌면서 overhead가 1% 미만
