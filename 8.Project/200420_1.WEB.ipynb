## 2.  웹 서버
- 각 언어마다 서버 언어를 처리하는 동작이 각기 상이함

#### 웹 서버
- 인터넷을 통해서 요청된 웹 컨텐츠의 전달을 도와주는 하드웨어와 소프트웨어를 말함.

> 기본적으로 정적
- 클라이언트가 HTTP 오청을 하면 그 리소스를 그대로 보내주는게 웹 서버의 역할

> 동적 처리
- 웹 서버에서 어플리케이션을 작동시키기 위한 인터페이스가 필요함
- CGI, WAS,WSGI, FCGI

#### CGI(Common Gateway Interface)
- 웹 서버에서 어플리케이션을 작동시키기 위한 인터페이스
- 정적인 웹서버를 동적으로 가능하게 하기 위해 등장
- 웹 서버에 내장함으로써 따로 프로세스를 생성하여 외부 플그램 실행없이 내부에서 전부 처리
- 방식: 요청 -> 웹서버(아파치 등)-> 웹서버가 직접 실행 프로그램
- perl, c/ c++에서 많이 처리
- 원론적인 CGI는 많이 사용하지 않음

#### WAS(Wep Application Server)
- 웹 서버가 동적으로 기능하면 WAS
- thrbahdls ruddndpsms
- 웹 서버 위에 서버 어플리케이션을 얹어놓은 것인 WAS
- 방식: 요청 -? 웹서버 -> 웹어플리케이션 서버(톰캣 등) -> 프로그램


#### WSGI
- 파이썬에서 어플리케이션, 파이선 스크립트(웹 어플리케이션)가 웹 서버와 통신하기 위한 인터페이스
- 프로토콜 개념으로 이해도 가능
- 서버와 앱 양단으로 나뉘어져 있음.
- WSGI 리퀘스트를 처리하려면 서버에서 환경정보와 콜백 함수를 제공
- 앱을 그 요청을 처리하고 콜백함수를 통해 서버에 응담
- 방식: 요청-> 웹서버-> WSGI SERGER(middleware)-> WSGI를 지원하는 웹 어플리케이션( Django, flask 등)

##### Flask
- 파이썬의 마이크로프레임워크
- WSGI미들웨어와 템플릿엔진으로구성ㅇ되어 있음.
- 핵심기능만 간결하게 유지하고 확장이 가능한 마이크로 프레임워크
- 파이썬에서 web server, was, wsgi미들웨어에 대한 제약이 없음.

##### Django
- 보안이 우수하고, 유이보수가 편리한 웹사이트를 신속하게 개발하고 하도로 도움을 줌

> 장점
- comlite(모든것이 개발 가능)
- 다용도 개발 가능
- 보안성, 확장성, 유지보수가 쉬움. 특정 서버 플랫폼에 국한되지 않음.

### 텐서플로우 서빙
- 만들어진 모델은 운영환경에서 사용할 수 있도록 도와줌
- 텐서플로우버싱 API를 이용해서 입력 데이터를 반영

#### Nginx
- 트랙픽이 많은 웹사이트를 위해 확장성을 위해 설꼐한 비동기 이벤트 기반 구조의 웹서버
