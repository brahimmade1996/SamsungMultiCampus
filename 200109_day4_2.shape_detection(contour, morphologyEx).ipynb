# 5. countour: 점을 찾은 개념 -> approximation 
# => 꼭지점 찾기

```python
img = cv2.imread('./images/shape3.png', 1)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
thr, mask = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

mask = 255-mask

contours,_ = cv2.findContours(image = mask, 
                              mode = cv2.RETR_EXTERNAL, # 최외곽선만 찾기 (도넛모양의 경우 부분적으로 찾을 수 있음.)
                              method = cv2.CHAIN_APPROX_SIMPLE)  # 
```

```python
# contour의 길이를 구하는 함수
con = contours[2]

peri = cv2.arcLength(con, 
                     True)  # contour가 열려있을 때 폐곡선으로 처리해라

apprx = cv2.approxPolyDP(curve= con, 
                         epsilon= 0.02*peri,  # approximation 하는 정도 -> 실수값
                         closed=True)

# epsiln: 점의 길이, shape의 크기에 따라 모두 달라짐 
#-> 길이의 2%의 변화가 생기면 appproximation해라
```

```python
cv2.drawContours(image =img, 
                 contours =apprx,
                 contourIdx =-1, # countour가 여러개 있을 때 모두 그려줘
                 color =(255,0,0), 
                 thickness =10)
imshow('', img)
```

## 5.1. 모든 도형에 approximation적용

```python
img = cv2.imread('./images/shape3.png', 1)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
thr, mask = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

mask = 255-mask

contours,_ = cv2.findContours(image = mask, 
                              mode = cv2.RETR_EXTERNAL, # 최외곽선만 찾기 (도넛모양의 경우 부분적으로 찾을 수 있음.)
                              method = cv2.CHAIN_APPROX_SIMPLE)  # 
```

```python
for con in contours:
    peri = cv2.arcLength(con, 
                     True)  # contour가 열려있을 때 폐곡선으로 처리해라

    approx = cv2.approxPolyDP(curve= con, 
                         epsilon= 0.02*peri,  # approximation 하는 정도 -> 실수값
                         closed=True)
    
    print('shape=',len(con),'lenth=',peri, 'approx=',len(approx))
          
    p = tuple(con[1][0])
    cv2.drawContours(img, [approx], -1,color =(255,0,0), thickness =10)
    imshow('', img)
```

#### drawContours할 때
- approx:로 넣으면 점만 찍어서 그려주기
- [approx]로 넣으면 외곽선을 모두 긂

## 5.2. 모든 도형에 approximation적용  + 글자쓰기

```python
img = cv2.imread('./images/shape3.png', 1)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
thr, mask = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

mask = 255-mask

contours,_ = cv2.findContours(image = mask, 
                              mode = cv2.RETR_EXTERNAL, # 최외곽선만 찾기 (도넛모양의 경우 부분적으로 찾을 수 있음.)
                              method = cv2.CHAIN_APPROX_SIMPLE)  # 
```

```python
for i,con in enumerate(contours):
    
    # contour의 길이
    peri = cv2.arcLength(con, 
                     True)  # contour가 열려있을 때 폐곡선으로 처리해라

    # contour의 넓이 : 넓이가 너무 작으면 노이즈 -> 노이즈 제거용도
    area = cv2.contourArea(con)
    approx = cv2.approxPolyDP(curve= con, 
                         epsilon= 0.02*peri,  # approximation 하는 정도 -> 실수값
                         closed=True)
    
    print('shape=',len(con),'lenth=',peri, 'approx=',len(approx))
          
    p = (con[i][0][0]+10, con[i][0][1]+10)
    font = cv2.FONT_HERSHEY_SIMPLEX
    if len(approx) ==3:
        cv2.putText(img, 'Triangle', p, font, 1, (0))
    elif len(approx) == 4:
        cv2.putText(img, 'Rectangle', p, font, 1, (0))
    elif len(approx) == 5:
        cv2.putText(img, 'Pentagon', p, font, 1, (0))
    elif len(approx) < 15:
        cv2.putText(img, 'Ellipse', p, font, 1, (0))
    else:
        cv2.putText(img, 'Circle', p, font, 1, (0))

  
    cv2.drawContours(img, [approx], -1,color =(255,0,0), thickness =10)
    imshow('', img)
```

- 잘 선택되었지만, 모두 1개로 인식. 모두 붙어있는 것이 문제

## 6.1. 침식/ 팽창  => 흰색 기준으로 침식/ 팽창이 이루어짐!

> 침식-> 개체가 작아져서 분리됨
- 약한 연결고리가 있는 부분이 끊어지는 효과

> 팽창
- 작은 구멍들이 매워지는 효과

```python
# 팽창
dilated = cv2.morphologyEx(src = binary,
                         op =cv2.MORPH_DILATE,    # 팽창: cv2.MORPH_ERODE,
                         kernel = (3,3),  # 9개 픽셀 중 하나만 선택하겠다
                         iterations = 20)


# 침식
eroded = cv2.morphologyEx(src = binary,
                         op =cv2.MORPH_ERODE,
                         kernel = (3,3),  # 9개 픽셀 중 하나만 선택하겠다
                         iterations = 10)

imshow('', np.hstack([binary, dilated, eroded]))
```

## 6.2. open/close

- 둘 다 원래 영상 크기는 어느정도 유지 됨.

> open
- 침식 후 팽창
- 뾰족한 부분 이 있을 때 전체적인 이미지가 줄어들고, 팽창되면 뾰족한 부분들이 없어짐.

> close
- 팽창 후 침식
- 팽창하면서 구멍들이 메워진 후, 다시 작아짐.

```python
# open: 침식 후 팽창
opened = cv2.morphologyEx(binary,
                         cv2.MORPH_OPEN,    # 팽창: cv2.MORPH_ERODE,
                         cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5)),
                           100)


# closed: 팽창 후 침식
closed = cv2.morphologyEx( binary,
                         cv2.MORPH_CLOSE,    
                          cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5)),
                         100)

imshow('', np.hstack([binary, opened, closed]))
```

## 6.3. coin 문제 풀어보자

```python
frame = cv2.imread('./images/coin.png',1)
roi = frame.copy()
gray= cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
```


### 6.3.1. 노이즈 제거 -> 블러링

```python
gray_blur = cv2.GaussianBlur(gray, (7,7), 0)
imshow('', gray_blur)
```

### 6.3.2. binary

```python
thresh = cv2.adaptiveThreshold(gray_blur, 255, 
                              cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                              cv2.THRESH_BINARY_INV ,  # 배경과 전경 바꾸기(255-maskd와 같음.)
                              45,6)
imshow('',thresh)
```

> 도넛부분
- 전경인데, 배경으로 인식해버림

### 6.3.3. 구멍 매꾸기: closed 연산

```python
# kernel직접 만들기
kernel = np.ones((11,11), np.uint8)


# 노이즈 없애기
closed = cv2.morphologyEx(thresh,
                         cv2.MORPH_CLOSE,    
                         kernel,
                         )

imshow('',  np.hstack([thresh,closed]))
```

### 6.3.4 외관선 찾기: contour

```python
cont_img = closed.copy()
contours, _ = cv2.findContours(cont_img, 
                               cv2.RETR_EXTERNAL,  # 최외곽선만 찾기
                              cv2.CHAIN_APPROX_SIMPLE)

for cnt in contours:
    
    # 너무 작거나 너무 큰 면적은 무시
    area = cv2.contourArea(cnt)
    if area < 2000 or area > 4000:
        continue
        
    # 원이 아닌 경우 무시   
    if len(cnt) < 5:
        continue
        
    # 주어진 점에 대해 타원으로 만들기
    ellipse = cv2.fitEllipse(cnt)
    cv2.ellipse(roi, ellipse, (0,255,0), 2)
    
imshow('out', thresh)
imshow('out', cont_img)
imshow('out', roi)
```

## 6.5. 영수증 외곽선 찾기- 강사님꺼

```python
img = cv2.imread('./images/mart2.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
gray = cv2.medianBlur(gray, 21)
thr, mask = cv2.threshold(gray, 0, 255,
        cv2.THRESH_BINARY+cv2.THRESH_OTSU)
imshow("", mask)

kernel = np.ones((5, 5), np.uint8)           
closing = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel, iterations=5)   
imshow("", closing)                                 

contours, _ = cv2.findContours(closing, cv2.RETR_EXTERNAL,
cv2.CHAIN_APPROX_SIMPLE)

print(len(contours))    

print(  f"shape = {len(contours)}"  )


colors = [(255,0,0), (0,255,0), (0,0,255), (255,255,0), (255,0,255)]

maxArea = 0

for i in range(len(contours)) :        
    con = contours[i]
    peri = cv2.arcLength(con, True)
    area = cv2.contourArea(con)  
    approx = cv2.approxPolyDP(con, 0.02 * peri, True)    
    print(  f"shape={len(con)}  length={peri}  approx={len(approx)}   area={area}"  )    
    if area > maxArea  :
        maxArea = area
        maxContour = approx
    
cv2.drawContours(img, [maxContour], -1, (0,0,255), 20)
   
    
imshow("", img)
cv2.imwrite("output.jpg", img)
```
